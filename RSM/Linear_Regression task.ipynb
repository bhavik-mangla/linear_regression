# Linear Regression using Gradient Descent

```python
def mse(slope, x_train, y_train):
    tot_error = 0
    pred = []

    pred.append(np.dot(x_train,slope))
    tot_error += (y_train - (np.dot(x_train,slope)))** 2
    
    error = sum(tot_error) / (2*float(len(x_train)))

    
    return error, pred
```

```python
def rsq(prediction, y_test):
    #
    total_data = len(prediction)
    #Average of total prediction 
    y_avg = np.sum(y_test)/total_data
    #total sum of square error
    tot_err = np.sum((y_test-y_avg)**2)
    #total sum of squared error of residuals
    res_err = np.sum((y_test-prediction)**2)
    #
    r2 = 1 - (res_err / tot_err)
    return r2
```

```python
#gradient descent 
def grad_descent(s_slope, l_rate, iter_val, x_train, y_train):
    
    final_slope = s_slope

    for i in range(iter_val):
        
        n_data = x_train.shape[1] 

        int_slope = np.ones((n_data,1)) 
        # print(y_train.shape,x_train.shape,s_slope.shape,np.dot( x_train,s_slope).shape ,
        # np.transpose(x_train).shape,(y_train - (np.dot( x_train,s_slope))).shape)

        # print(i)
        int_slope+= - (2/n_data) * np.dot(np.transpose(x_train), (y_train - (np.dot( x_train,s_slope))))
            
        final_slope = final_slope - (l_rate * int_slope)

        
        
    return  final_slope
```

**if i try with iterations greater than this, it makes all slope values NaN**

```python
#defining slope and intercept value as 0 
learning_rate = 0.00001

n_data = x_train.shape[1]

start_slope = np.ones((n_data+1,1))
bias = np.ones(( x_train.shape[0],1))
x_train_b = np.append(bias, x_train, axis=1)
bias1 = np.ones(( x_test.shape[0],1))
x_test_b = np.append(bias1, x_test, axis=1)


iteration = 10000
#intial run
grad_slope = grad_descent(start_slope, learning_rate, iteration, x_train_b, y_train)
final_e_value, prediction = mse(grad_slope, x_test_b, y_test)
#
prediction = np.array(prediction)
prediction = np.array(prediction).reshape(200,1)
print(prediction.shape,"\n\n")

for i in range(200):
    print(prediction[i],y_test[i])

print(final_e_value)


print(rsq(y_test, prediction))
print(metrics.r2_score(y_test, prediction))
```
